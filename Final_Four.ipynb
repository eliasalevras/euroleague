## MAKE BOXSCORE LIKE EUROLEAGUE BOXSCORE

```python
import pandas as pd

# =========================================
# LOAD RAW EUROLEAGUE BOXSCORE
# =========================================

# input_path = r"C:/Users/alevr/Desktop/website_project/data/A404_2025.csv"
input_path = r"C:/Users/alevr/Desktop/website_project/data/A405_2025.csv"


df = pd.read_csv(input_path)

# =========================================
# KEEP ONLY REAL PLAYERS
# =========================================

# Remove team / total rows
df = df[
    (~df["Player"].astype(str).str.contains("Team", na=False)) &
    (~df["Player"].astype(str).str.contains("Total", na=False))
].copy()

# =========================================
# CREATE CLEAN WEBSITE BOXSCORE
# =========================================

website_boxscore = pd.DataFrame({
    "Team": df["Team"],
    "Dorsal":df["Dorsal"],
    "Player": df["Player"],

    "Min": df["Minutes"],

    "PTS": df["Points"],

    # =========================
    # 2FG
    # =========================
    "2FG_M": df["FieldGoalsMade2"],
    "2FG_A": df["FieldGoalsAttempted2"],

    # =========================
    # 3FG
    # =========================
    "3FG_M": df["FieldGoalsMade3"],
    "3FG_A": df["FieldGoalsAttempted3"],

    # =========================
    # FT
    # =========================
    "FT_M": df["FreeThrowsMade"],
    "FT_A": df["FreeThrowsAttempted"],

    # =========================
    # REBOUNDS
    # =========================
    "OREB": df["OffensiveRebounds"],
    "DREB": df["DefensiveRebounds"],
    "REB": df["TotalRebounds"],

    # =========================
    # OTHER STATS
    # =========================
    "AST": df["Assistances"],
    "STL": df["Steals"],
    "TO": df["Turnovers"],

    "BLK_F": df["BlocksFavour"],
    "BLK_A": df["BlocksAgainst"],

    "FOULS_C": df["FoulsCommited"],
    "FOULS_R": df["FoulsReceived"],

    "PIR": df["Valuation"],

    "+/-": df["Plusminus"]
})

# =========================================
# PERCENTAGES
# =========================================

website_boxscore["2FG_%"] = (
    website_boxscore["2FG_M"] /
    website_boxscore["2FG_A"].replace(0, 1)
    * 100
).round(1)

website_boxscore["3FG_%"] = (
    website_boxscore["3FG_M"] /
    website_boxscore["3FG_A"].replace(0, 1)
    * 100
).round(1)

website_boxscore["FT_%"] = (
    website_boxscore["FT_M"] /
    website_boxscore["FT_A"].replace(0, 1)
    * 100
).round(1)

# =========================================
# SAVE
# =========================================

# output_path = r"C:/Users/alevr/Desktop/website_project/data/sf1_boxscore.csv"
output_path = r"C:/Users/alevr/Desktop/website_project/data/sf2_boxscore.csv"


website_boxscore.to_csv(
    output_path,
    index=False,
    encoding="utf-8-sig"
)

print("Saved:")
print(output_path)

print("\nPreview:")
print(website_boxscore.head())
```

```python
import pandas as pd

# input_path = "C:/Users/alevr/Desktop/website_project/data/semifinal_1.csv"
# output_path = "C:/Users/alevr/Desktop/website_project/data/Assist_Network_Summary_G404.csv"
# details_output_path = "C:/Users/alevr/Desktop/website_project/data/Assist_To_Scorer_Details_G404.csv"



input_path = "C:/Users/alevr/Desktop/website_project/data/semifinal_2.csv"
output_path = "C:/Users/alevr/Desktop/website_project/data/Assist_Network_Summary_G405.csv"
details_output_path = "C:/Users/alevr/Desktop/website_project/data/Assist_To_Scorer_Details_G405.csv"

df = pd.read_csv(input_path)

# Sort σωστά τη ροή του play-by-play
df = df.sort_values(["Gamecode", "TRUE_NUMBEROFPLAY"]).reset_index(drop=True)

made_shots = ["2FGM", "3FGM"]

details = []

for i in range(len(df) - 1):
    shot_row = df.iloc[i]
    next_row = df.iloc[i + 1]

    # Made shot και αμέσως μετά assist
    if (
        shot_row["PLAYTYPE"] in made_shots
        and next_row["PLAYTYPE"] == "AS"
        and shot_row["Gamecode"] == next_row["Gamecode"]
        and shot_row["CODETEAM"] == next_row["CODETEAM"]
    ):
        points = 2 if shot_row["PLAYTYPE"] == "2FGM" else 3

        details.append({
            "Gamecode": shot_row["Gamecode"],
            "TEAM_CODE": shot_row["CODETEAM"],
            "TEAM": shot_row["TEAM"],

            "ASSIST_PLAYER_ID": next_row["PLAYER_ID"],
            "ASSIST_PLAYER": next_row["PLAYER"],

            "SCORER_PLAYER_ID": shot_row["PLAYER_ID"],
            "SCORER_PLAYER": shot_row["PLAYER"],

            "SHOT_TYPE": shot_row["PLAYTYPE"],
            "ASSISTED_POINTS": points,

            "PERIOD": shot_row["PERIOD"],
            "MINUTE": shot_row["MINUTE"],
            "MARKERTIME": shot_row["MARKERTIME"]
        })

assist_details = pd.DataFrame(details)

assist_summary = (
    assist_details
    .groupby(
        [
            "TEAM_CODE",
            "TEAM",
            "ASSIST_PLAYER_ID",
            "ASSIST_PLAYER",
            "SCORER_PLAYER_ID",
            "SCORER_PLAYER"
        ],
        as_index=False
    )
    .agg(
        ASSISTS=("ASSISTED_POINTS", "count"),
        ASSISTED_POINTS=("ASSISTED_POINTS", "sum")
    )
    .sort_values(["TEAM_CODE", "ASSISTED_POINTS"], ascending=[True, False])
)

assist_details.to_csv(details_output_path, index=False, encoding="utf-8-sig")
assist_summary.to_csv(output_path, index=False, encoding="utf-8-sig")

print("Assist details:")
print(assist_details.head(20).to_string())

print("\nAssist network summary:")
print(assist_summary.to_string())

print("\nSaved:")
print(details_output_path)
print(output_path)
```
